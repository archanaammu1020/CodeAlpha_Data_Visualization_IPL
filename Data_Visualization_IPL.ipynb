```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
matches = pd.read_csv('/content/sample_data/matches.csv')
deliveries = pd.read_csv('/content/sample_data/deliveries.csv')
```

```python
plt.figure(figsize=(12,6))

matches['winner'].value_counts().plot(kind='bar')

plt.title('Most Winning Teams in IPL')
plt.xlabel('Teams')
plt.ylabel('Wins')

plt.show()
```

```python
matches['toss_decision'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Toss Decision Analysis')
plt.ylabel('')

plt.show()
```

```python
plt.figure(figsize=(12,6))

matches['player_of_match'].value_counts().head(10).plot(kind='bar')

plt.title('Top Player of the Match Winners')
plt.xlabel('Players')
plt.ylabel('Awards')

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(x='Season', data=matches)

plt.title('Matches Played Per Season')

plt.xticks(rotation=45)

plt.show()
```

```python
top_teams = deliveries.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)

top_teams.plot(kind='bar', figsize=(12,6))

plt.title('Top Run Scoring Teams')
plt.xlabel('Teams')
plt.ylabel('Runs')

plt.show()
```
